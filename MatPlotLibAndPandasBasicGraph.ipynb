```python
import pandas as pd
```

```python
from matplotlib import pyplot as mlp
```

```python
# define two data list for test plotting data
a=[1,2,3]
b=[4,5,6]
mlp.plot(a,b) # Plot Graph
mlp.xlabel('a axis') # Set X axis label
mlp.ylabel('b axis') # Set Y axis label
mlp.title('Test Graph') # Set Title
mlp.show() # Show Graph
```

```python
nsedata = pd.read_csv('data.csv')
```

```python
nsedata.head()
```

```python
nsedata['Date'] = pd.to_datetime(nsedata['Date']) # This is to convert datatype of column 
mlp.plot(nsedata.Date,nsedata.Close)
mlp.xlabel('Dates')
mlp.ylabel('Close')
mlp.title('NSE Jan Data')
```

```python
# Now plotting opena and closing in same graph
mlp.plot(nsedata.Date,nsedata.Open)
mlp.plot(nsedata.Date,nsedata.Close)
mlp.xlabel('Dates')
mlp.ylabel('Close/Open')
mlp.title('NSE Jan Data')
mlp.legend(['open','close'])
```
